```python
import rasterio
import matplotlib.pyplot as plt

# Caminho do arquivo .tif
tif_path = "../outputs/2_Sentinel1_converted_dB_2.tif"
# Abre o arquivo com rasterio
with rasterio.open(tif_path) as src:
    # Lê a primeira banda (ou especifique outra)
    band1 = src.read(1)
    
    # Exibe informações do raster
    print("CRS:", src.crs)
    print("Dimensões (linhas, colunas):", src.height, src.width)
    print("Número de bandas:", src.count)
    
    # Plota a banda
    plt.figure(figsize=(8, 6))
    plt.imshow(band1, cmap="gray")
    plt.colorbar(label="Valor do pixel")
    plt.title("Imagem GeoTIFF - Banda 1")
    plt.xlabel("Coluna")
    plt.ylabel("Linha")
    plt.show()
```
